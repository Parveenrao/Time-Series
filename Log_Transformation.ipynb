```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data with exponential growth
data = [10, 20, 40, 80, 160, 320]
series = pd.Series(data)

# Apply log transformation
log_series = np.log(series)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(series, label='Original Series', marker='o')
plt.plot(log_series, label='Log Transformed', marker='o', color='red')
plt.title("Log Transformation Effect")
plt.legend()
plt.show()
```
