# Simple EXP Smothing


```python
import pandas as pd
import matplotlib.pyplot as plt
alpha = 0.3

data = pd.Series([10, 12, 13, 15, 14, 16])

ses = data.ewm(alpha = 0.3 , adjust = False).mean()

ses
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data, marker='o', label='Original Data')
plt.plot(ses, marker='o', label=f'SES (alpha={alpha})', color='red')
plt.title("Simple Exponential Smoothing")
plt.legend()
plt.show()
```

#  Double EXP Smoothing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
data = pd.Series([10, 12, 13, 15, 14, 16, 18, 20, 23, 25])

model = ExponentialSmoothing(data , trend = 'add' , seasonal = None)

fit  = model.fit(smoothing_level = 0.3 , smoothing_slope = 0.2)

des = fit.fittedvalues
```

```python
# plot


plt.figure(figsize = (6,6))
plt.plot(data , marker = 'o' , label  = 'Original Data' )
plt.plot(des , marker = 'o' , label = 'Double Exponential Smoothing')
plt.title("Double Exponential Smoothing (Holt Winters Methods")
plt.legend()
plt.show()
```

# Holt Winter Exponential Smoothing

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Sample data with trend + seasonality
data = [30, 21, 29, 31, 40, 32, 41, 43, 50, 42, 52, 54,
        60, 51, 61, 63, 70, 62, 72, 74, 80, 71, 81, 83]
series = pd.Series(data)

# Holt-Winters model
model = ExponentialSmoothing(series, trend='add', seasonal='add', seasonal_periods=4)
fit = model.fit(smoothing_level=0.3, smoothing_slope=0.2, smoothing_seasonal=0.4)
hw = fit.fittedvalues

# Plot
plt.figure(figsize=(10, 5))
plt.plot(series, marker='o', label='Original Data')
plt.plot(hw, marker='o', label='Holt-Winters Smoothed', color='red')
plt.title("Holt-Winters Triple Exponential Smoothing")
plt.legend()
plt.show()
```
