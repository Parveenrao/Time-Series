```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a 10-year monthly time series
date_range = pd.date_range(start="2010-01-01", periods=120, freq='M')

# ---- Components ----
# 1. Trend: steady upward slope
trend = 0.5 * np.arange(len(date_range))

# 2. Seasonality: yearly repeating pattern
seasonality = 10 * np.sin(2 * np.pi * (date_range.month) / 12)

# 3. Cyclic variation: long-term business cycle (~7 years, not fixed)
cyclic = 15 * np.sin(2 * np.pi * date_range.year / 7)

# 4. Irregular (noise): random unpredictable variations
noise = np.random.normal(0, 3, len(date_range))

# Combine all to make the full time series
time_series = trend + seasonality + cyclic + noise

# ---- Plot all components ----
plt.figure(figsize=(14, 10))

plt.subplot(5, 1, 1)
plt.plot(date_range, time_series, color='black')
plt.title("Complete Time Series")

plt.subplot(5, 1, 2)
plt.plot(date_range, trend, color='blue')
plt.title("Trend Component")

plt.subplot(5, 1, 3)
plt.plot(date_range, seasonality, color='green')
plt.title("Seasonal Component")

plt.subplot(5, 1, 4)
plt.plot(date_range, cyclic, color='purple')
plt.title("Cyclic Variation")

plt.subplot(5, 1, 5)
plt.plot(date_range, noise, color='red')
plt.title("Irregular (Noise) Component")

plt.tight_layout()
plt.show()
```
