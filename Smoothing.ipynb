```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# simple time series 

series = pd.Series([10 ,11,12, 14, 15, 17, 18, 19 ,20 ,21 ,22])
```

```python
# simple moving averages(window_szie = 3)

sma = series.rolling(window = 3).mean()
```

```python
# plotting all the component

plt.figure(figsize = (6, 6))
plt.plot(series, label = 'Original Data' , color = 'Red')
plt.plot(sma  , label = 'Simple Moving Average(k = 3)' , color = 'Blue')
plt.title("Simple Moving Averages Method")
plt.legend()
plt.show()
```

# Weighted MOving Averages ( Assign more weight to recent observations)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.Series([0, 12, 13, 15, 14, 16, 18, 17, 19, 20, 22, 21])

weight = np.array([1 ,2 ,3])   # for more weight to recent observation

wma = series.rolling(window = 3).apply(lambda x : np.dot(x , weight) / weight.sum() , raw = True)

                   


                   
```

```python
# plot the data

plt.figure(figsize = (6,6))
plt.plot(series , label = 'Original Data' , color = 'Red')
plt.plot(wma , label = 'Weighted Moving Averages' , color = 'Blue')
plt.title("Weighted Moving Average (WMA)")
plt.legend()
plt.show()
```
