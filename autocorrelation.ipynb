```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
time_series = pd.Series(0.5 * np.arange(n) + np.random.normal(0 , 2 ,n ))
```

```python
plt.figure(figsize=(10, 4))
plt.plot(time_series)
plt.title("Sample Time Series")
plt.show()
```

```python
from statsmodels.tsa.stattools import acf
```

```python
autocorr_values = acf(time_series, nlags=100)
print("Autocorrelation values:\n", autocorr_values)
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(time_series, lags=20)
plt.title("Autocorrelation Function (ACF)")
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(time_series, lags=20, method='ywm')
plt.title("Partial Autocorrelation Function (PACF)")
plt.show()
```
