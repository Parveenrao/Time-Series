```python
import pandas as pd
import numpy as np
import statsmodels
import matplotlib.pyplot as plt
```

```python
data = pd.Series([100 ,105 , 111, 120 ,130, 141])
```

```python
# first order differencing

diff1 = data.diff()

diff1
```

```python
# plot

plt.figure(figsize = (6,6))
plt.plot(data , label = 'Original Series' , marker = 'o' , color = 'Red' )
plt.plot(diff1 , label = 'First order Differencing' , marker = 'o' ,color =  'Blue')
plt.title("First Order Differencing" )
plt.legend()
plt.show()
```

# 2nd Order Differencing

```python
data = pd.Series([100 ,105 , 111, 120 ,130, 141])


diff1 = data.diff()

diff2 = diff1.diff()
```

```python
diff2
```

```python
#plot

plt.figure(figsize = (6,6))
plt.plot(data , label = 'Original Series' , marker = 'o' , color = 'Red' )
plt.plot(diff1 , label = 'First order Differencing' , marker = 'o' ,color =  'Blue')
plt.plot(diff2, label = 'Second order Differencing' , marker = 'o' ,color =  'Green')
plt.title("Second Order Differencing" )
plt.legend()
plt.show()
```

# Seasonal Differencing

```python
import pandas as pd
import matplotlib.pyplot as plt

# Monthly data with seasonality (2 years, m=12)
data = [120, 150, 170, 180, 160, 140, 130, 125, 140, 160, 180, 200,
        130, 160, 180, 190, 170, 150, 140, 135, 150, 170, 190, 210]
series = pd.Series(data)

# Seasonal differencing (m = 12)
seasonal_diff = series.diff(12)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(series, label='Original Series')
plt.plot(seasonal_diff, label='Seasonally Differenced Series', color='red')
plt.title("Seasonal Differencing (m=12)")
plt.legend()
plt.show()
```
